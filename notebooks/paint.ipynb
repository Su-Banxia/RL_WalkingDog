```python
import matplotlib.pyplot as plt
import pandas as pd

# 读取 CSV
csv_path = "../models/training_rewards.csv"
df = df = pd.read_csv(csv_path)

# 画图
df['avg_reward_smooth'] = df['avg_reward'].rolling(window=10).mean()

plt.figure(figsize=(8,5))
plt.plot(df['timestep'], df['avg_reward'], marker='o', linestyle='-', alpha=0.3, label='Raw')
plt.plot(df['timestep'], df['avg_reward_smooth'], color='red', linewidth=2, label='Smoothed')
plt.legend()
plt.title('TD3 Training: Avg Reward vs Timestep')
plt.xlabel('Timestep')
plt.ylabel('Average Reward')
plt.grid(True)
plt.tight_layout()
plt.show()

# 保存图像（可选）
# plt.savefig('training_curve.png')

plt.show()
```
